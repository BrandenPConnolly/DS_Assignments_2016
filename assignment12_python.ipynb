```python
class MagicCard(object):
    
    def __init__(
                self, 
                name="", 
                manaCost = "", 
                cmc=0, 
                colors=None, 
                type="", 
                supertypes=None, 
                types=None, 
                subtypes=None, 
                rarity="", 
                text="", 
                flavor="", 
                artist="",
                number="",
                power="",
                toughness="", 
                layout="",
                multiverseid=0,
                imageName="",
                id=""
                ):
        self.name = name
        self.manaCost = manaCost
        self.cmc = cmc
        self.colors = colors
        self.type = type
        self.supertypes = supertypes
        self.subtypes = subtypes
        self.rarity = rarity
        self.text = text
        self.flavor = flavor
        self.artist = artist
        self.number = number
        self.power = power
        self.toughness = toughness
        self.layout = layout
        self.multiverseid = multiverseid
        self.imageName = imageName
        self.id = id
    
    def get_name(self):
        # return name of card
        return self.name
    
    def get_manaCost(self):
        # return manaCost of card
        return self.manaCost
    
    def get_cmc(self):
        # return cmc of card
        return self.cmc
    
    def get_colors(self):
        # return colors of card
        return self.colors

    def get_type(self):
        # return type of card
        return self.type
    
    def get_supertypes(self):
        # return supertypes of card
        return self.supertypes
    
    def get_subtypes(self):
        # return subtypes of card
        return self.subtypes
    
    def get_rarity(self):
        # return rarity of card
        return self.rarity

    def get_text(self):
        # return text of card
        return self.text
    
    def get_flavor(self):
        # return manaCost of card
        return self.flavor
    
    def get_artist(self):
        # return artist of card
        return self.artist
    
    def get_number(self):
        # return number of card
        return self.number

    def get_power(self):
        # return power of card
        return self.power
    
    def get_toughness(self):
        #return toughness of card
        return self.toughness
    
    def get_layout(self):
        #return layout of card
        return self.layout
    
    def get_multiverseid(self):
        # return multiverseid of card
        return self.multiverseid

    def get_imageName(self):
        # return imageName of card
        return self.imageName
    
    def get_id(self):
        # return id of card
        return self.id
```
